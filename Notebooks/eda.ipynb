```python
# Import necessary libraries
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# Load the dataset
file_path = r"C:\Users\anura\Desktop\Tobacco Morality Prediction and Analysis\Notebooks\output.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())
print("\nDataset Head:")
print(df.head())
```

```python
# 1. Generate a Pandas Profiling Report
print("\nGenerating Pandas Profiling Report...")
profile = ProfileReport(
    df,
    title="Tobacco Use and Mortality: EDA Report",
    explorative=True,
)
```

```python
# Save the report to an HTML file
output_path = r"C:\Users\anura\Desktop\Tobacco Morality Prediction and Analysis\EDA Report"
profile.to_file(output_path)
print(f"\nEDA Report saved at: {output_path}")
```

```python
# 2. Additional EDA using Pandas
print("\nBasic Descriptive Statistics:")
print(df.describe())
```

```python
print("\nChecking for Missing Values:")
print(df.isnull().sum())
```

```python
# Apply backfill to the specific column
df['Normalized Varenicline Cost'].fillna(method='bfill', inplace=True)
```

```python
print("\nChecking for Missing Values:")
print(df.isnull().sum())
```

```python
# 3. Correlation Analysis
print("\nCorrelation Matrix:")
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
# Optional: Visualize the correlation matrix using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 4. Analyze Trends Over Time
# Plotting Tobacco Affordability and Smoking Prevalence trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="Year", y="Tobacco Affordability", label="Tobacco Affordability", marker="o")
sns.lineplot(data=df, x="Year", y="Smoking Prevalence", label="Smoking Prevalence", marker="o")
plt.title("Trends Over Time: Tobacco Affordability vs Smoking Prevalence")
plt.xlabel("Year")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# 5. Distribution of Smoking Rates by Age Groups
age_columns = ["16-24", "25-34", "35-49", "50-59", "60 and Over"]

plt.figure(figsize=(12, 6))
for age_group in age_columns:
    sns.lineplot(data=df, x="Year", y=age_group, label=f"{age_group} Smoking Rate", marker="o")
plt.title("Smoking Rates by Age Group")
plt.xlabel("Year")
plt.ylabel("Smoking Rate")
plt.legend()
plt.grid(True)
plt.show()
```
